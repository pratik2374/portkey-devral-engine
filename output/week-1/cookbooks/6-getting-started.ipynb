## Step 2: Setup & Imports

We need to initialize the Portkey client. Never hardcode API keys! We'll pull them from environment variables.

```python
import os
from portkey_ai import Portkey

# Set your API keys as environment variables or define them here for the notebook demo
os.environ["PORTKEY_API_KEY"] = "your-portkey-api-key"   # Replace with your key
os.environ["OPENAI_API_KEY"] = "your-openai-api-key"     # Replace with your key
os.environ["ANTHROPIC_API_KEY"] = "your-anthropic-api-key" # Replace with your key

PORTKEY_API_KEY = os.getenv("PORTKEY_API_KEY")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if PORTKEY_API_KEY == "your-portkey-api-key":
    print("⚠️ Warning: Please replace placeholder API keys above.")

# Initialize Portkey client routing to OpenAI by default
client = Portkey(
    api_key=PORTKEY_API_KEY,
    provider="openai",
    Authorization=OPENAI_API_KEY
)

print("✅ Portkey client initialized!")
```

## Step 3: Your First Chat Completion

Making a request through Portkey looks exactly like making a standard OpenAI API call. The difference? Your request is now routed through Portkey's gateway, unlocking observability, caching, and guardrails instantly.

```python
try:
    # Make the request just like you would with OpenAI
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": "Explain the concept of an AI Gateway in one short sentence."}]
    )
    
    result = response.choices[0].message.content
    print(f"✅ OpenAI Response says: {result}")

except Exception as e:
    print(f"❌ Error: {e}")
    print("💡 Tips:")
    print("  - Check your PORTKEY_API_KEY is valid")
    print("  - Verify your provider API key is set")
    print("  - Ensure the model name is correct")
```

## 🔄 Switch Providers in One Line

One of Portkey's superpowers — switch between 250+ LLM providers without changing your prompt structure or core logic:

Let's switch from OpenAI to Anthropic's Claude 3.5 Sonnet.

```python
try:
    # Switch to Anthropic — just change provider and auth during initialization
    anthropic_client = Portkey(
        api_key=PORTKEY_API_KEY,
        provider="anthropic",
        Authorization=os.getenv("ANTHROPIC_API_KEY")
    )
    
    # Same API call structure, totally different provider behind the scenes
    response = anthropic_client.chat.completions.create(
        model="claude-3-5-sonnet-20241022",
        messages=[{"role": "user", "content": "Explain the concept of an AI Gateway in one short sentence."}]
    )
    
    print(f"✨ Anthropic Response says: {response.choices[0].message.content}")
    
except Exception as e:
    print(f"❌ Error: {e}")
    print("Did you set your Anthropic API Key?")
```

## Bonus: The 2-Line OpenAI Migration

If you already have a massive codebase built entirely on the `openai` Python SDK, you don't need to rewrite anything. You can route everything through Portkey by changing just two lines of code when defining the OpenAI client.

```python
from openai import OpenAI
from portkey_ai import PORTKEY_GATEWAY_URL, createHeaders

try:
    # Initialize standard OpenAI client...
    migration_client = OpenAI(
        api_key=os.getenv("OPENAI_API_KEY"),
        base_url=PORTKEY_GATEWAY_URL,            # ← Line 1: Change Base URL
        default_headers=createHeaders(            # ← Line 2: Add Portkey headers
            api_key=PORTKEY_API_KEY,
            provider="openai"
        )
    )
    
    # Existing deep application logic remains exactly the same!
    response = migration_client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": "Is this routed through Portkey? (Respond with YES or NO)"}]
    )
    
    print(f"Migration Client Response: {response.choices[0].message.content}")

except Exception as e:
    print(f"❌ Error: {e}")
```
